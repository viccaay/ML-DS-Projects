```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/noshowappointments.csv")
```

```python
data['show']=data['No-show']
data.head()
```

```python
data.corr(method ='pearson')
```

```python
#Age and Hypertension show correlation 
sns.barplot(data=data, x='Hipertension', y='Age', hue='show', palette='PuRd')
```

```python
#Age and Diabetes show correlation 
sns.barplot(data=data, x='Diabetes', y='Age', hue='show')
```

```python
#Do people with Diabetes have Hypertension as well?
sns.barplot(data=data, x='Diabetes',y='Hipertension', hue='Gender')
```

```python
#showing up vs SMS reception for reminder 
sns.barplot(data=data, x='show', y='Age', hue='SMS_received')
```
